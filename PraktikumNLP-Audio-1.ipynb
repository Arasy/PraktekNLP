$$ y = f(t) = A \ sin(\omega t) \ = A \ sin(2 \pi ft)$$

- $A$ = amplitudo
- $f$ = frekuensi
- $t$ = indeks waktu
- $\omega = 2 \pi f$ = frekuensi angular

### Fungsi Sinyal Gabungan

Lawan dari monotone (**nada tunggal**) adalah polytone (**nada jamak**). Yang pernah menggunakan HP tahun 2000an mungkin merasakan perubahan dari ringtone *monophonic* ke ringtone *poliphonic*. Konsepnya sama. Politone adalah gabungan dari beberapa frekuensi yang berbeda yang dimainkan bersamaan. 

Fungsi gabungannya adalah hasil penjumlahan dari fungsi-fungsi sinus.

$$y_m = \Sigma (\ A_i \sin(2 \pi f_i t)\ ) = A_1 \ sin(2 \pi f_1 t) + A_2 \ sin(2 \pi f_2 t) + A_3 \ sin(2 \pi f_3 t) +\ ... $$

- $A_i$ = amplitudo untuk frekuensi ke-i
- $f_i$ = frekuensi ke-i
- $t$ = indeks waktu

misalkan ada 3 gelombang
- gelombang 1 dengan frekwensi 130 Hz dan amplitudo 10
- gelombang 2 dengan frekwensi 196 Hz dan amplitudo 8
- gelombang 3 dengan frekwensi 220 Hz dan amplitudo 6

Gelombang ini bisa kita buat dalam sebuah variabel sebagai berikut :

```python
A1 = 10
f1 = 130
A2 = 8
f2 = 196
A3 = 6
f3 = 220

y_gab = A1 * np.sin(2 * np.pi * f1 * t) +  A2 * np.sin(2 * np.pi * f2 * t) + A3 * np.sin(2 * np.pi * f3 * t)

plt.plot(y_gab[:1000])
plt.show()

sd.play(y_gab,fs)
```

# Transformasi Fourier

Transformasi Fourier akan mengubah gelombang dari domain waktu ke domain frekwensi


sumber gambar : https://analyticsvidhya.com

Dengan mengaplikasikan formula berikut ke fungsi gelombangnya

$$F(k) = \frac{1}{T}\ \int_{-\frac{T}{2}}^{\frac{T}{2}} f(t)\ e^{-2\pi\ i\ \frac{k}{T}\ t}\ dt$$
where :<br>

- $f(t)$ : fungsi gelombang dalam domain waktu
- $e$ : Euler's number 2.718281828459....
- $k$ : indeks frekwensi
- $i$ : bilangan imajiner $(\sqrt{-1})$
- $T$ : periode

```python
#transformasi fourier diterapkan ke gelombang monotone (nada tunggal)
plt.plot(y[:1000])
Y = np.abs(np.fft.fft(y))
plt.plot(Y[:1000])
```

```python
#transformasi fourier diterapkan ke gelombang polytone (nada jamak) 

Y = np.abs(np.fft.fft(y_gab))
plt.plot(Y[:1000])
```

```python
#transformasi fourier diterapkan ke gelombang suara

Y = np.abs(np.fft.fft(y_wicara))
plt.plot(Y[0:np.int(Y.size/2)]) 
```

# Short Time Fourier Transform

Apabila suatu gelombang wicara ditransformasi menggunakan transformasi fourier secara langsung akan menghilangkan informasi jangka pendek karena indeks waktu (t) berubah ke indeks frekuensi (k). Padahal informasi jangka pendek ini sangat bermanfaat jika digabungkan dengan informasi spektralnya. Karena itu muncullah metode yang dikenal dengan **Short Time Fourier Transform** (STFT). 

Daripada mengubah keseluruhan sinyalnya sekaligus, sinyal ini dibagi menjadi bagian-bagian kecil yang saling overlapping dan ditransformasikan bagian per bagian. 



STFT memiliki sebuah konstanta untuk memisahkan setiap bagiannya (chunk) yang disebut sebagai jarak lompatan (hopping distance). Setiap lompatan diberi bobot menggunakan **window function** agar transisi satu bagian ke bagian sebelahnya bisa digabungkan kembali. Ada banyak **window function** yang dapat digunakan dan masing-masing memiliki karakteristik yang unik. Pada praktikum ini, fungsi yang digunakan adalah **Hamming Window** yang umum digunakan.

Hamming Window dinyatakan sebagai berikut
$$H[n] = 0.54 - 0.46\ cos \left[ \frac{2\pi n}{N - 1} \right]$$

#### Chunk
Misalkan kita punya sinyal, kita dapat ambil sebagian sinyal tersebut (chunk) dalam frame waktu yang ditentukan. Ukuran inilah yang kita sebut dengan frame size.

```python
#generate sinyal
fs = 48000                   # frekuensi sampel

t = np.linspace(0, 1, fs)    # sampel
A = 1                        # unit
f = 220                      # Hz

y = A * np.sin(2 * np.pi * f * t)

#ambil sebagian dari sinyal tersebut
N = 1000                     # frame size
chunk = y[:N]

#plot
plt.plot(chunk)
plt.show
```

#### Hamming Window
Kita buat fungsi jendelanya (*window function*)

```python
H = 0.54 - 0.46 *  np.cos(2 * np.pi * np.arange(0, N) / (N - 1))
plt.plot(H)
```

lalu kita terapkan window function tersebut ke sinyal yang kita miliki

```python
frame = np.multiply(chunk, H)
plt.plot(frame)
plt.plot(H)
```

### Penerapan STFT ke Sinyal Wicara
Sekarang kita terapkan ke file audio yang kita punya. File contoh.wav memiliki frekuensi sampel 48.000 Hz, dengan durasi sekitar 4 detik. Kita akan buat frame sizenya 0.2 detik (9.600 data). 

```python
plt.plot(y_wicara)
```

```python
#credit from Tirtadwipa Manunggal (https://github.com/linerocks)
def STFT(x, N = 512, overlap = 0.5) :
    if overlap > 0.5 :
        overlap = 0.5
    elif overlap < 0.1 :
        overlap = 0.1
    
    # Calculating frames
    hop_size     = np.int( np.floor(N * (1 - overlap)) )
    total_frames = np.int( np.ceil(x.size / hop_size) )
    
    # Zero padding
    x = np.append( x, np.zeros((total_frames + 1) * hop_size - x.size) )
    
    # STFT container
    result = np.empty( (total_frames, np.int(N/2)), dtype=np.float32 )
    
    # Window
    H = 0.54 - 0.46 *  np.cos(np.arange(0, N) * 2 * np.pi / (N - 1))
        
    # Building
    for i in range(total_frames) :
        hop_index    = i * hop_size
        chunk        = x[hop_index : hop_index + N]
        frame        = np.multiply(chunk, H)
        spectrum     = np.abs(np.fft.fft(frame))
        result[i, :] = spectrum[0:np.int(N/2)]
    
    return result   
```

```python
N= 9600

spectrum = np.log(STFT(y_wicara,N))
plt.imshow(spectrum.T, origin='lower', aspect='auto', extent=[0, spectrum.size / fs, 0, 8000])
plt.title('Spectrogram')
plt.colorbar()
```
